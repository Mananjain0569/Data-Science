```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel('toughestsport.xlsx')
print(data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

spd_sport = data.groupby('SPORT')['SPD'].mean().sort_values(ascending=False)

# Visualize the results
plt.figure(figsize=(26, 8))
sns.barplot(x=spd_sport.index, y=spd_sport.values)
plt.title('Average Speed Rate by Sport')
plt.xlabel('Sport')
plt.ylabel('Average Speed Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
Speed_agility_correlation = data['SPD'].corr(data['AGI'])

print(f"Correlation between Speed and Agility: {Speed_agility_correlation}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the Matplotlib figure
plt.figure(figsize=(20, 10))

# Scatter plot with regression line
ax = sns.regplot(x='FLX', y='AGI', data=data, scatter_kws={'s': data['AGI']*15},
                 line_kws={'color': 'red', 'linestyle': '--'}, ci=None)

# Define and highlight outliers (customize the criteria based on your data)
some_threshold = 75  # Placeholder value, replace with your own threshold
outliers = data[(data['FLX'] > some_threshold) | (data['AGI'] > some_threshold)]

# Highlight outliers with a different marker and color
ax.scatter(x=outliers['FLX'], y=outliers['AGI'], marker='o', color='orange', s=outliers['AGI']*15)

# Annotate every 4th point with the sport name for better readability
for i in range(0, len(data), 1):
    plt.annotate(data['SPORT'][i], (data['FLX'][i], data['AGI'][i]),
                 textcoords="offset points", xytext=(0, 5), ha='center')

# Customize the plot
plt.title('Flexibility vs Agility in Different Sports')
plt.xlabel('Flexibility Score')
plt.ylabel('Agility Score')
plt.grid(True)
plt.show()
```

```python
# Calculate correlation between strength and rankings
strength_ranking_correlation = data['STR'].corr(data['RANK'])

# Print the correlation value
print(f"Correlation between Strength and Rankings: {strength_ranking_correlation}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate mean and confidence interval for Strength
strength_stats = data.groupby('SPORT')['STR'].agg(['mean', 'std'])

# Sort the sports by mean strength
strength_stats = strength_stats.sort_values(by='mean', ascending=True)

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='mean', y=strength_stats.index, data=strength_stats, errorbar ='sd', palette='viridis')

# Adding labels and title
plt.title('Mean Strength with Confidence Intervals Across Different Sports', fontsize=16)
plt.xlabel('Mean Strength', fontsize=14)
plt.ylabel('Sports', fontsize=14)

# Customize the axis labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Remove the right and top spines for aesthetics
sns.despine()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Example data
data = pd.DataFrame({
    'Boxing': np.random.rand(100),
    'Ice Hockey': np.random.rand(100)
})

x = data['Boxing']
y = data['Ice Hockey']

rng = np.random.RandomState(0)
colors = rng.rand(len(data))
sizes = 1000 * rng.rand(len(data))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
plt.xlabel('Boxing')
plt.ylabel('Ice Hockey')
plt.title('Scatter Plot')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
movies = pd.read_excel("/content/toughestsport.xlsx")

# Remove the "%" symbol and convert "Rotten Tomatoes" to float
#movies['BOXING'] = movies["BOXING"].str.replace("%", "").astype(float)

# Drop non-numeric columns
movies_numeric = movies.select_dtypes(include=['float64', 'int64'])

# Calculate the correlations
correlations = movies_numeric.corr(method='pearson')

# Generate the heatmap
sns.heatmap(correlations, cmap="coolwarm")
plt.show()
```
